```python
import numpy as np
import matplotlib.pyplot as plt
import time
import math


def newton_raphson(f, g, xi, xj, eps,max_iter):
    """
    Implementasi metode Newton-Raphson.

    Args:
        f (function): Fungsi f(x).
        g (function): Turunan pertama f(x), yaitu g(x).
        xi (float): Nilai awal akar.
        eps (float): Kesalahan absolut (epsilon).

    Returns:
        tuple: (akar, nilai f(x) pada akar, jumlah iterasi).
    """
    iteration = 0  # Jumlah iterasi
    while iteration < max_iter:

        xi = xj - f(xj) / g(xj)  # Iterasi Newton-Raphson
        print(f"Iterasi {iteration+1})")
        iteration += 1
        if abs(f(xi)) <= eps:  # Kondisi konvergensi
            return xi, f(xi), iteration # Changed 'n' to 'iteration'
    # If the loop completes without finding a root, return None or raise an exception
    # Here, we return None with the current values
    return xi, f(xi), iteration

f = lambda x: x**2 - 3  # Fungsi
g = lambda x: 2 * x     # Turunan
xi = 1.0
xj = 2.0  # Nilai awal
eps = 1e-6  # Toleransi kesalahan

root, fx_value, iterations = newton_raphson(f, g, xi, xj, eps,1000)
print(f"Akar: {root}")
print(f"Nilai f(x) pada akar: {fx_value}")
print(f"Jumlah iterasi: {iterations}")
```

```python
def newton_raphson_max_iter(f, g, xi, xj, eps, max_iter):
    """
    Implementasi metode Newton-Raphson dengan batas iterasi maksimum.

    Args:
        f (function): Fungsi f(x).
        g (function): Turunan pertama f(x), yaitu g(x).
        xi (float): Nilai awal akar.
        eps (float): Kesalahan absolut (epsilon).
        max_iter (int): Jumlah iterasi maksimum.

    Returns:
        tuple: (akar, nilai f(x) pada akar, jumlah iterasi, berhasil/tidak).
    """
    n = 0  # Jumlah iterasi
    while n < max_iter:
        n += 1
        xi = xj - f(xj) / g(xj)  # Iterasi Newton-Raphson
        if abs(f(xi)) <= eps:  # Kondisi konvergensi
            return xi, f(xi), n, True  # Berhasil menemukan akar
    return xi, f(xi), n, False
    #return None, None, n, False  # Tidak konvergen dalam iterasi maksimum

# Contoh penggunaan
if __name__ == "__main__":
    import math

    # Definisi fungsi f(x) dan turunan g(x)
    f = lambda x: x**2 - 3  # Fungsi
    g = lambda x: 2 * x     # Turunan

    xi = 1.0
    xj = 2.0       # Nilai awal
    eps = 1e-6     # Toleransi kesalahan
    max_iter = 100  # Iterasi maksimum

    root, fx_value, iterations, success = newton_raphson_max_iter(f, g, xi, xj, eps, max_iter)
    if success:
        print(f"Akar: {root}")
        print(f"Nilai f(x) pada akar: {fx_value}")
        print(f"Jumlah iterasi: {iterations}")
    else:
        print(f"Metode gagal menemukan akar dalam {iterations} iterasi.")
```

```python
# Python3 implementation of Bisection
# Method for solving equations

MAX_ITER = 1000

# An example function whose solution
# is determined using Regular Falsi Method.
# The function is x^3 - x^2 + 2


# Prints root of func(x) in interval [a, b]
def regulaFalsi(f, a, b, tol, max_iter):
    if f(a) * f(b) >= 0:
        print("You have not assumed right a and b")
        return -1

     # Initialize result
    iteration = 0
    root_list = []
    while (b-a) / 2.0 > tol and iteration < max_iter:
        # Find the point that touches x axis
        xr = (a * f(b) - b * f(a))/ (f(b) - f(a))
        print(f"Iterasi {iteration+1}: Lower Bound = {a}, Upper Bound = {b}")
        root_list.append(xr)

        # Check if the above found point is root
        if f(xr) == 0:
            return xr, root_list,iteration # Return root and root_list

        # Decide the side to repeat the steps
        elif f(xr) * f(a) < 0:
            b = xr
        else:
            a = xr
        iteration += 1

    # Return value of root is not found
    #print("The value of root is : " , '%.4f' %c)
    return xr, root_list,iteration # Return root and root_list outside the while loop


start_time = time.time()
root, root_list,iteration_list = regulaFalsi(f, 1, 2, 1e-14, 100)
end_time = time.time()
cpu_time = end_time - start_time

print("\n=== Regula Falsi Method ===")
print("Found root:", root)
print("Average CPU Time:", cpu_time, "seconds")
print("Number of iterations:", iteration_list)

# This code is contributed by "Sharad_Bhardwaj".
```

```python


# Fungsi Regula Falsi
def regulaFalsi(f, a, b, tol, max_iter):
    if f(a) * f(b) >= 0:
        print("You have not assumed right a and b")
        return -1

    # Initialize result
    iteration = 0
    root_list = []

    while iteration < max_iter:
        # Find the point that touches x axis
        xr = (a * f(b) - b * f(a)) / (f(b) - f(a))
        print(f"Iterasi {iteration+1}: Lower Bound = {a}, Upper Bound = {b}, Approximate Root = {xr}")
        root_list.append(xr)

        # Check if the above found point is root or close enough
        if abs(f(xr)) < tol:
            return xr, root_list, iteration  # Root found with acceptable tolerance

        # Decide the side to repeat the steps
        elif f(xr) * f(a) < 0:
            b = xr
        else:
            a = xr

        iteration += 1

    # Return the root and root_list after max iterations or tolerance met
    return xr, root_list, iteration

# Mulai pengukuran waktu
start_time = time.time()

# Panggil fungsi Regula Falsi, pastikan iterasi berhenti pada 34
root, root_list, iteration_list = regulaFalsi(f, -2, 4, 1e-14, 100)

# Akhir pengukuran waktu
end_time = time.time()

# Hitung waktu CPU rata-rata
cpu_time = end_time - start_time

# Cetak hasil
print("\n=== Regula Falsi Method ===")
print("Found root:", root)
print("Average CPU Time:", cpu_time, "seconds")
print("Number of iterations:", iteration_list)
```
