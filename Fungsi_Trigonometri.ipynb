```python
import matplotlib.pyplot as plt

# Data masalah (problems) dan iterasi untuk setiap metode
problems = ['F_8', 'F_9', 'F_10','F_13','F15']
iteration_bisection = [46, 47, 49, 46, 45]
iteration_trisection = [30, 30, 31, 30, 29 ]
iteration_regula = [13, 7, 100, 44, 16]
iteration_newton = [7, 7, 14, 6, 9]
iteration_hybrid_bisection_regula = [5, 4, 7, 6, 5]
iteration_hybrid_trisection_regula = [7, 5, 8, 7, 4]

# Membuat plot
plt.figure(figsize=(10, 6))

# Plot untuk setiap metode
plt.plot(problems, iteration_bisection, linestyle='--', marker='o', color='blue', label='Bisection')
plt.plot(problems, iteration_trisection, linestyle='--', marker='o', color='orange', label='Trisection')
#plt.plot(problems, iteration_regula, linestyle='--', marker='o', color='green', label='Regula Falsi')
#plt.plot(problems, iteration_newton, linestyle='--', marker='o', color='red', label='Newton-Raphson')
plt.plot(problems, iteration_hybrid_bisection_regula, linestyle='-', marker='o', color='purple', label='Hybrid Bisection-Regula Falsi')
#plt.plot(problems, iteration_hybrid_trisection_regula, linestyle='--', marker='o', color='brown', label='Hybrid Trisection-Regula Falsi')

# Menambahkan label pada setiap titik data
for i, value in enumerate(iteration_bisection):
    plt.annotate(f"{value}", (problems[i], iteration_bisection[i]), textcoords="offset points",  xytext=(0, 5), ha='center', fontsize=8, color='blue')

for i, value in enumerate(iteration_trisection):
    plt.annotate(f"{value}", (problems[i], iteration_trisection[i]), textcoords="offset points",  xytext=(0, 5), ha='center', fontsize=8, color='orange')

#for i, value in enumerate(iteration_regula):
 #   plt.annotate(f"{value}", (problems[i], iteration_regula[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8, color='green')

#for i, value in enumerate(iteration_newton):
 #   plt.annotate(f"{value}", (problems[i], iteration_newton[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8, color='red')

for i, value in enumerate(iteration_hybrid_bisection_regula):
    plt.annotate(f"{value}", (problems[i], iteration_hybrid_bisection_regula[i]), textcoords="offset points",  xytext=(0, 5), ha='center', fontsize=8, color='purple')

#for i, value in enumerate(iteration_hybrid_trisection_regula):
   # plt.annotate(f"{value}", (problems[i], iteration_hybrid_trisection_regula[i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8, color='brown')

# Menambahkan judul dan label sumbu
plt.title('Perbandingan jumlah iterasi metode algoritma Bisection,Trisection dan hybrid bisection-regula falsi Fungsi Trigonometri', fontsize=16, pad=20)
plt.xlabel('Fungsi(f)', fontsize=14, labelpad=15)
plt.ylabel('iterasi', fontsize=14, labelpad=15)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Menyesuaikan layout
plt.tight_layout()

# Menampilkan plot
plt.show()
```
